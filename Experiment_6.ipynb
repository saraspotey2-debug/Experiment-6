```python
#6.9 Salary Calculation with Allowances.
# Write a python program to calculate the gross salary of an employee.
# If basic<=10000: HRA = 20%, DA = 80%.
#If basic <= 20000: HRA = 25%, DA = 90%.
#Otherwise: HA = 30%, DA=95%.
basic=float(input("Enter basic salary: "))
if basic<=10000:
   hra=basic * 0.20
   da = basic * 0.80
elif basic <= 20000:
   hra= basic * 0.25
   da=basic * 0.90
else:
   hra= basic * 0.30
   da= basic * 0.95
gross= basic+hra+da
print("Gross Salary:",gross)
```
